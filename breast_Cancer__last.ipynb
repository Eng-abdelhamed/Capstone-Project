1-**Metastasis**

*   Malignant cells grow in an uncontrolled way and can invade nearby tissues and spread to other parts of the body through the blood and lymph system.

2 -**Menopause**

*   The time of life when a woman's ovaries stop producing hormones and menstrual periods stop
Hormonal imbalance such as
continuous exposure to estrogen leads to high risk of
breast cancers. In this scenario, women with natural
menopause are more likely to prone to develop
cancers as twice as high because of hormonal
disturbances. Hence, based on Menopause attribute
breast cancer cells are detected and classified into
benign and malignant

3-**Inv-Nodes**


*   INV Nodes are the number (range 0 -
39) of axillary lymph nodes that contain metastatic
breast cancer visible on histological examination.
Hence INV Nodes are taken as an attribute to detect
and classify cancer cells as benign and malignant.





**Benign** -> a condition, tumor, or growth that is not cancerous

**Malignant** -> A term used to describe cancer. Malignant cells grow in an uncontrolled way and can invade nearby tissues and spread to other parts of the body through the blood and lymph system.

# Relationship Between Menopause and Disease Severity

```python
from scipy.stats import chi2_contingency
menopause_result_ct = pd.crosstab(df['Menopause'], df['Result'])
chi2, p, dof, expected = chi2_contingency(menopause_result_ct)
print(f"Chi-Square Statistic: {chi2:.2f}")
print(f"P-Value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
```

# Key Factors Influencing Breast Cancer Risk

```python
numeric_df = df.select_dtypes(include=['number'])
correlations = numeric_df.corr()['Result'].sort_values(ascending=False)
correlations
```

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

X = df[['Age', 'TumorSize', 'Inv-Nodes', 'Menopause', 'Metastasis']]
y = df['Result']

model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

feature_importance = model.feature_importances_

plt.figure(figsize=(8, 5))
plt.bar(X.columns, feature_importance, color='teal')
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.title("Feature Importance in Breast Cancer Prediction")
plt.xticks(rotation=45)
plt.show()

for feature, importance in zip(X.columns, feature_importance):
    print(f"{feature}: {importance:.4f}")
```

```python
# functions that check if malignant based tumor size see if it hifg risk or low ( < 5 ="low" , >=5 = "high")
def check_for_risk(row):
    if row["Result"] == 0:
        return 0
    return 1 if row["TumorSize"] < 5 else 2


## add these results as y (label needed output)
df["y"] = df.apply(check_for_risk, axis=1)
```

```python
df['y'].value_counts()
```

```python
# import hot encoding to encode the y
from sklearn.preprocessing import OneHotEncoder
```

```python
# encoding the result for each row
OneHotEncoder(sparse_output=False).fit_transform(df[["y"]])
```

```python
# put it into varible to pass it as new columns in new df ( carries only the results)
result = OneHotEncoder(sparse_output=False).fit_transform(df[["y"]]).astype(int) # change the type from float to
```

```python
result
```

```python
# new df
result_df = pd.DataFrame(result, columns=["Benign", "Malignant Low Risk", "Malignant High Risk"])
```

```python
# drop unneeded cols
df.drop(columns=["y", "Result"], inplace=True)
```

```python
processed_df = pd.concat([df, result_df], axis=1)
processed_df.head()
```
